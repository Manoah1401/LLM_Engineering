```python
from openai import OpenAI
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama =OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')
```

```python
email = """
Dear Sir/Madam,



I am writing to formally request the issuance of my degree certificate and to authorize my friend, Mr. Isaac Babu, to collect it on my behalf, as I am unable to come in person due to personal reasons.

I am submitting all the required documents along with this email for your reference. This includes:

a) A copy of my college ID card

b) An authorization letter addressed to your office, granting Mr. Issac Babu the right to collect the certificate on my behalf

I kindly request you to assist him in collecting the document at your earliest convenience. Please let me know if any further information or documentation is required.



Thank you



Regards, 

Manoah Noble

URK21CS1173

manoahnob@gmail.com

9321764317

"""
```

```python
system_prompt="""
You are a snarky and inexperienced email writing assistant that analyzes the contents of an email,
provides a snarky and humurous subject titles and also modifies and makes the email funny but dont make spelling mistakes or grammatical mistakes.
Do not give me anything else. Just reply with the subject line and email.
"""
```

```python
user_prompt= """
Here are the contents of an email.
Provide a subject title for this email.
Also refine the email if there are any mistakes and give me the message content along with the subject title

"""
```

```python
messages=[{"role":"system", "content":system_prompt},{"role":"user", "content":user_prompt+email}]
```

```python
responses=ollama.chat.completions.create(model="llama3.2",messages=messages)
```

```python
from IPython.display import Markdown, display
display(Markdown(responses.choices[0].message.content))
```
